````python
from pathlib import Path
from paddleocr import PPStructureV3

input_file = "/content/raw_page-0002.jpg"
output_path = Path("./output")

pipeline = PPStructureV3(
  use_doc_orientation_classify=True,
  use_doc_unwarping=True,
  use_textline_orientation=True,
  device="gpu"
)
output = pipeline.predict(input=input_file)

markdown_list = []
markdown_images = []

for res in output:
    md_info = res.markdown
    markdown_list.append(md_info)
    markdown_images.append(md_info.get("markdown_images", {}))

markdown_texts = pipeline.concatenate_markdown_pages(markdown_list)

mkd_file_path = output_path / f"{Path(input_file).stem}.md"
mkd_file_path.parent.mkdir(parents=True, exist_ok=True)

with open(mkd_file_path, "w", encoding="utf-8") as f:
    f.write(markdown_texts)

for item in markdown_images:
    if item:
        for path, image in item.items():
            file_path = output_path / path
            file_path.parent.mkdir(parents=True, exist_ok=True)
            image.save(file_path)
````

````python
import re
import json
import io
import base64
import requests
from pathlib import Path
import sys
from typing import List, Dict, Any, Optional, Tuple
import markdown
from markdown.extensions import codehilite, fenced_code, tables, toc, nl2br, sane_lists, extra
from docx import Document
from docx.shared import Pt, RGBColor, Inches
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT
from docx.enum.style import WD_STYLE_TYPE
from docx.oxml import OxmlElement
from docx.oxml.ns import qn
from bs4 import BeautifulSoup, NavigableString, Tag

# For Pygments (install with: pip install Pygments)
from pygments import highlight
from pygments.lexers import get_lexer_by_name, guess_lexer
from pygments.formatters import HtmlFormatter
from pygments.util import ClassNotFound


class MarkdownToDocxConverter:
    """Converts Markdown content to a DOCX file while preserving structure and formatting."""

    def __init__(self):
        self.doc: Optional[Document] = None
        self.pygments_formatter = HtmlFormatter(noclasses=True)

    def _add_or_get_style(self, style_name: str, style_type: WD_STYLE_TYPE):
        """
        Adds a new style if it doesn't exist, or returns the existing one.
        This prevents the "document already contains style" error.
        """
        try:
            return self.doc.styles[style_name]
        except KeyError:
            return self.doc.styles.add_style(style_name, style_type)

    def create_styles(self):
        """Creates custom styles for the document if they don't already exist."""
        if self.doc is None:
            raise ValueError("Document object is not initialized. Call convert() first.")

        # Style for code blocks
        code_style = self._add_or_get_style('CodeBlock', WD_STYLE_TYPE.PARAGRAPH)
        code_style.font.name = 'Consolas'
        code_style.font.size = Pt(9)
        code_style.paragraph_format.space_before = Pt(6)
        code_style.paragraph_format.space_after = Pt(6)
        code_style.paragraph_format.left_indent = Inches(0.3)

        # Apply shading only if it's not already set (checking existence is hard, so we just set)
        pPr = code_style.element.get_or_add_pPr()
        shading = pPr.find(qn('w:shd'))
        if shading is None: # Add shading if it doesn't exist
            shading = OxmlElement('w:shd')
            shading.set(qn('w:val'), 'clear')
            shading.set(qn('w:color'), 'auto')
            shading.set(qn('w:fill'), 'F0F0F0') # Light gray fill color
            pPr.append(shading)


        # Style for inline code
        inline_code_style = self._add_or_get_style('InlineCode', WD_STYLE_TYPE.CHARACTER)
        inline_code_style.font.name = 'Consolas'
        inline_code_style.font.size = Pt(9)
        inline_code_style.font.color.rgb = RGBColor(205, 78, 0) # Darker Red/Orange for inline code

        rPr = inline_code_style.element.get_or_add_rPr()
        inline_shading = rPr.find(qn('w:shd'))
        if inline_shading is None: # Add shading if it doesn't exist
            inline_shading = OxmlElement('w:shd')
            inline_shading.set(qn('w:val'), 'clear')
            inline_shading.set(qn('w:color'), 'auto')
            inline_shading.set(qn('w:fill'), 'E0E0E0') # Light gray fill for inline code
            rPr.append(inline_shading)


        # Styles for headings (H1 to H6)
        for i in range(1, 7):
            heading_style = self.doc.styles[f'Heading {i}']
            heading_style.font.color.rgb = RGBColor(0, 0, 0) # Black color
            heading_style.font.bold = True
            heading_style.font.size = Pt(26 - i * 2) # Decreasing font size for lower heading levels

        # Style for blockquotes
        quote_style = self._add_or_get_style('Quote', WD_STYLE_TYPE.PARAGRAPH)
        quote_style.font.name = 'Calibri'
        quote_style.font.size = Pt(10)
        quote_style.font.italic = True
        quote_style.paragraph_format.left_indent = Inches(0.5)
        quote_style.paragraph_format.space_before = Pt(8)
        quote_style.paragraph_format.space_after = Pt(8)
        quote_style.paragraph_format.keep_together = True

        pPr = quote_style.element.get_or_add_pPr()
        pBdr = pPr.find(qn('w:pBdr'))
        if pBdr is None: # Add border if it doesn't exist
            pBdr = OxmlElement('w:pBdr')
            leftBdr = OxmlElement('w:left')
            leftBdr.set(qn('w:val'), 'single')
            leftBdr.set(qn('w:sz'), '12') # 1.5 pt
            leftBdr.set(qn('w:space'), '10')
            leftBdr.set(qn('w:color'), 'ADD8E6')
            pBdr.append(leftBdr)
            pPr.append(pBdr)


    def add_code_block(self, code_text: str, language: str = ''):
        """Adds a code block with syntax highlighting (using Pygments) to the document."""
        p = self.doc.add_paragraph()
        p.style = 'CodeBlock'

        try:
            lexer = get_lexer_by_name(language)
        except ClassNotFound:
            try:
                lexer = guess_lexer(code_text)
            except ClassNotFound:
                lexer = get_lexer_by_name('text')

        highlighted_html = highlight(code_text, lexer, self.pygments_formatter)
        soup = BeautifulSoup(highlighted_html, 'html.parser')
        code_tag = soup.find('code')
        if code_tag:
            for content in code_tag.contents:
                if isinstance(content, NavigableString):
                    p.add_run(str(content))
                elif isinstance(content, Tag):
                    run = p.add_run(content.get_text())
                    style_attr = content.get('style')
                    if style_attr:
                        color_match = re.search(r'color:\s*#([0-9a-fA-F]{6})', style_attr)
                        if color_match:
                            hex_color = color_match.group(1)
                            rgb = tuple(int(hex_color[i:i+2], 16) for i in (0, 2, 4))
                            run.font.color.rgb = RGBColor(*rgb)
                        if 'bold' in style_attr:
                            run.bold = True
                        if 'italic' in style_attr:
                            run.italic = True
                        if 'underline' in style_attr:
                            run.underline = True
        else:
            p.add_run(code_text)


    def process_inline_code(self, paragraph, text: str):
        """Handles inline code within a paragraph."""
        parts = re.split(r'`([^`]+)`', text)
        for i, part in enumerate(parts):
            if i % 2 == 0:
                if part:
                    paragraph.add_run(part)
            else:
                run = paragraph.add_run(part)
                run.style = 'InlineCode'


    def add_horizontal_rule(self):
        """Adds a proper horizontal rule to the document."""
        p = self.doc.add_paragraph()
        p.paragraph_format.space_before = Pt(12)
        p.paragraph_format.space_after = Pt(12)

        p_pr = p.paragraph_format._pPr
        # Remove any existing pBdr to prevent duplicates if somehow called multiple times
        for child in list(p_pr):
            if child.tag == qn('w:pBdr'):
                p_pr.remove(child)

        pBdr = OxmlElement('w:pBdr')
        bottomBdr = OxmlElement('w:bottom')
        bottomBdr.set(qn('w:val'), 'single')
        bottomBdr.set(qn('w:sz'), '12')
        bottomBdr.set(qn('w:space'), '1')
        bottomBdr.set(qn('w:color'), 'D3D3D3')
        pBdr.append(bottomBdr)
        p_pr.append(pBdr)


    def process_html_element(self, element, parent_paragraph=None):
        """Recursively processes HTML elements and adds them to the DOCX."""
        if isinstance(element, NavigableString):
            text = str(element)
            if parent_paragraph and text.strip():
                # Apply styles based on direct parent (strong, em, del)
                current_run = parent_paragraph.add_run(text)
                if element.parent and (element.parent.name == 'strong' or element.parent.name == 'b'):
                    current_run.bold = True
                if element.parent and (element.parent.name == 'em' or element.parent.name == 'i'):
                    current_run.italic = True
                if element.parent and (element.parent.name == 'del' or element.parent.name == 's'):
                    current_run.font.strike = True
            return

        if not isinstance(element, Tag):
            return

        # Special handling for elements that affect alignment or overall block structure
        if element.name == 'div' and element.get('style') == 'text-align: center;':
            # Create a new paragraph for centering if content is not just text
            # This is a heuristic; direct table centering is better
            p = self.doc.add_paragraph()
            p.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
            for child in element.children:
                # If the child is a table, add it centered directly
                if child.name == 'table':
                    self.add_table(child, center_table=True) # Pass centering hint
                else:
                    self.process_html_element(child, p)
            return # Div handled, skip generic child processing

        if element.name in ['h1', 'h2', 'h3', 'h4', 'h5', 'h6']:
            level = int(element.name[1])
            self.doc.add_heading(element.get_text(), level=level)
        elif element.name == 'p':
            img_children = element.find_all('img', recursive=False)
            text_content = element.get_text(strip=True)

            if len(img_children) == 1 and not text_content:
                self.process_html_element(img_children[0])
            elif text_content.startswith("%%MERMAID_DIAGRAM_") and text_content.endswith("%%"):
                pass # Handled at the top-level
            else:
                p = self.doc.add_paragraph()
                # Process children, handling inline code splits here
                full_text = ''.join(str(c) for c in element.contents)
                self.process_inline_code(p, full_text)

        elif element.name == 'img':
            self.add_image(element.get('src'))
        elif element.name == 'a':
            if parent_paragraph:
                self.add_hyperlink(parent_paragraph, element.get_text(), element.get('href'))
            else:
                p = self.doc.add_paragraph()
                self.add_hyperlink(p, element.get_text(), element.get('href'))
        elif element.name == 'pre':
            code_element = element.find('code')
            if code_element:
                classes = code_element.get('class', [])
                code_text = code_element.get_text()

                if 'mermaid' in classes or 'language-mermaid' in classes:
                    if not code_text.strip().startswith("%%MERMAID_DIAGRAM_"): # Avoid re-processing placeholders
                        self.add_mermaid_diagram(code_text)
                else:
                    language = ''
                    for cls in classes:
                        if cls.startswith('language-'):
                            language = cls.replace('language-', '')
                            break
                    self.add_code_block(code_text, language)
            else:
                self.add_code_block(element.get_text())
        elif element.name == 'code' and parent_paragraph:
            run = parent_paragraph.add_run(element.get_text())
            run.style = 'InlineCode'
        elif element.name == 'ul':
            self._process_list(element, 'List Bullet')
        elif element.name == 'ol':
            self._process_list(element, 'List Number')
        elif element.name == 'blockquote':
            p = self.doc.add_paragraph(style='Quote')
            for child in element.children:
                self.process_html_element(child, p)
        elif element.name == 'strong' or element.name == 'b':
            if parent_paragraph:
                # Recursively process children, the NavigableString handler will apply bold
                for child in element.children:
                    self.process_html_element(child, parent_paragraph)
        elif element.name == 'em' or element.name == 'i':
            if parent_paragraph:
                for child in element.children:
                    self.process_html_element(child, parent_paragraph)
        elif element.name == 'del' or element.name == 's':
            if parent_paragraph:
                for child in element.children:
                    self.process_html_element(child, parent_paragraph)
        elif element.name == 'table':
            self.add_table(element) # No centering hint by default for tables processed here
        elif element.name == 'hr':
            self.add_horizontal_rule()
        else:
            # Fallback: recursively process children for unhandled tags
            for child in element.children:
                self.process_html_element(child, parent_paragraph)

    def _process_list(self, list_element: Tag, style_name: str):
        """Helper to process ordered and unordered lists, handles nesting."""
        for li in list_element.find_all('li', recursive=False):
            p = self.doc.add_paragraph(style=style_name)
            # Process children of <li>, combining them into a single paragraph as much as possible
            # We need to reconstruct the content of the list item for inline code processing
            li_content_html = ''.join(str(child) for child in li.contents)
            soup_li_content = BeautifulSoup(li_content_html, 'html.parser')
            for content_part in soup_li_content.contents:
                if isinstance(content_part, NavigableString):
                    self.process_inline_code(p, str(content_part))
                elif isinstance(content_part, Tag):
                    if content_part.name == 'code': # Explicit inline code tag
                        run = p.add_run(content_part.get_text())
                        run.style = 'InlineCode'
                    elif content_part.name in ['strong', 'b', 'em', 'i', 'del', 's']:
                        # Handle nested formatting within list item
                        run = p.add_run(content_part.get_text())
                        if content_part.name in ['strong', 'b']: run.bold = True
                        if content_part.name in ['em', 'i']: run.italic = True
                        if content_part.name in ['del', 's']: run.font.strike = True
                    else:
                        # For other tags inside <li> (e.g., links, images), process normally
                        self.process_html_element(content_part, p)
            # Handle nested lists correctly
            for nested_list in li.find_all(['ul', 'ol'], recursive=False):
                self._process_list(nested_list, 'List Bullet' if nested_list.name == 'ul' else 'List Number')


    def add_image(self, src: str):
        """Downloads and inserts an image from a URL into the document, attempting to center it."""
        if not src:
            return
        try:
            if src.startswith("data:"):
                print(f"Warning: Data URI images are not supported for direct download: {src[:50]}...")
                self.doc.add_paragraph(f"[Image (Data URI) not supported: {src[:50]}]", style='Body Text')
                return

            response = requests.get(src, stream=True, timeout=10)
            response.raise_for_status()

            image_stream = io.BytesIO(response.content)
            p = self.doc.add_paragraph()
            run = p.add_run()
            run.add_picture(image_stream, width=Inches(6))
            p.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
        except requests.exceptions.Timeout:
            print(f"Error: Timeout downloading image {src}")
            self.doc.add_paragraph(f"[Image download timeout: {src}]", style='Body Text')
        except requests.exceptions.RequestException as e:
            print(f"Error downloading image {src}: {e}")
            self.doc.add_paragraph(f"[Image not found or inaccessible: {src}]", style='Body Text')
        except Exception as e:
            print(f"Unexpected error adding image {src}: {e}")
            self.doc.add_paragraph(f"[Unexpected image error: {src}]", style='Body Text')

    def add_mermaid_diagram(self, code: str):
        """Renders and inserts a Mermaid diagram via mermaid.ink service."""
        try:
            graphbytes = code.encode("utf-8")
            base64_bytes = base64.urlsafe_b64encode(graphbytes)
            base64_string = base64_bytes.decode("ascii")

            url = f"https://mermaid.ink/img/{base64_string}"

            response = requests.get(url, timeout=15)
            response.raise_for_status()

            image_stream = io.BytesIO(response.content)
            p = self.doc.add_paragraph()
            run = p.add_run()
            run.add_picture(image_stream, width=Inches(6))
            p.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER

        except requests.exceptions.Timeout:
            print(f"Error: Timeout rendering Mermaid diagram.")
            self.doc.add_paragraph("[Mermaid diagram rendering timeout]", style='Body Text')
        except requests.exceptions.RequestException as e:
            print(f"Error rendering Mermaid diagram: {e}")
            self.doc.add_paragraph(f"[Diagram rendering error: {e}]", style='Body Text')
        except Exception as e:
            print(f"Unexpected error processing Mermaid: {e}")
            self.doc.add_paragraph(f"[Unexpected Mermaid error: {e}]", style='Body Text')

    def add_table(self, table_element: Tag, center_table: bool = False):
        """Adds an HTML table element to the DOCX document, with optional centering."""
        rows = table_element.find_all('tr')
        if not rows:
            return

        # Calculate max columns based on first row or max cells in any row
        cols = 0
        if rows:
            for cell in rows[0].find_all(['td', 'th']):
                cols += int(cell.get('colspan', 1))
        if cols == 0: # Fallback for tables without explicit header/colspan
            for row_element in rows:
                cols = max(cols, len(row_element.find_all(['td', 'th'])))

        if cols == 0:
            return

        table = self.doc.add_table(rows=0, cols=cols)
        table.style = 'Table Grid'

        # Set table alignment if requested (usually for tables wrapped in a centered div)
        if center_table:
            table.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
            # To truly center a table, you might need to adjust preferred width or apply a style.
            # python-docx table centering isn't as straightforward as paragraph centering.
            # A common trick is to use an implicit table style or adjust its XML directly.
            # For this example, setting table.alignment is the best high-level option.

        # *** IMPORTANT NOTE ON ROWSPAN ***
        # python-docx's high-level API does not directly support HTML-style rowspan.
        # Implementing rowspan accurately requires manual manipulation of WordprocessingML XML,
        # which is outside the scope of this general Markdown converter.
        # Cells with rowspan will be rendered as new, unmerged cells in subsequent rows.
        # You will see the content of the rowspan cell only in its initial row.
        # This will NOT replicate your example table's structure perfectly for merged cells.
        print("Warning: Table contains 'rowspan' attributes. python-docx does not directly support 'rowspan' for HTML tables through its high-level API. The table layout may not be perfectly replicated.")

        for row_idx, row_element in enumerate(rows):
            doc_row = table.add_row()
            cells_in_html_row = row_element.find_all(['td', 'th'])
            current_doc_cell_idx = 0

            for html_cell_element in cells_in_html_row:
                colspan = int(html_cell_element.get('colspan', 1))
                # rowspan = int(html_cell_element.get('rowspan', 1)) # Still not directly handled

                # Find the next available cell in the doc_row, skipping cells potentially merged
                # from previous colspans in this row
                while current_doc_cell_idx < cols and current_doc_cell_idx < len(doc_row.cells) \
                      and doc_row.cells[current_doc_cell_idx].paragraphs[0].text.strip():
                    # This check is imperfect as it relies on text content, not actual merge status
                    current_doc_cell_idx += 1

                if current_doc_cell_idx < cols:
                    cell = doc_row.cells[current_doc_cell_idx]

                    # Merge cells for colspan if valid
                    if colspan > 1 and current_doc_cell_idx + colspan <= cols:
                        # Make sure not to merge beyond table bounds
                        end_cell = doc_row.cells[min(cols - 1, current_doc_cell_idx + colspan - 1)]
                        cell.merge(end_cell)

                    # Clear existing content if any (e.g., from an initial paragraph)
                    for para in list(cell.paragraphs):
                        cell._element.remove(para._element)

                    # Add new paragraph and process content
                    p = cell.add_paragraph()
                    # Process content of the cell, including inline formatting
                    cell_content_html = ''.join(str(c) for c in html_cell_element.contents)
                    soup_cell_content = BeautifulSoup(cell_content_html, 'html.parser')
                    for content_part in soup_cell_content.contents:
                        if isinstance(content_part, NavigableString):
                            self.process_inline_code(p, str(content_part))
                        elif isinstance(content_part, Tag):
                            if content_part.name == 'code':
                                run = p.add_run(content_part.get_text())
                                run.style = 'InlineCode'
                            elif content_part.name in ['strong', 'b', 'em', 'i', 'del', 's']:
                                run = p.add_run(content_part.get_text())
                                if content_part.name in ['strong', 'b']: run.bold = True
                                if content_part.name in ['em', 'i']: run.italic = True
                                if content_part.name in ['del', 's']: run.font.strike = True
                            else:
                                self.process_html_element(content_part, p)

                    if html_cell_element.name == 'th':
                        for paragraph in cell.paragraphs:
                            for run in paragraph.runs:
                                run.bold = True

                    # Move past the merged cells
                    current_doc_cell_idx += colspan
                else:
                    print(f"Warning: Cell in row {row_idx} at HTML index {cells_in_html_row.index(html_cell_element)} could not be placed due to table structure/merge complexities.")


    def add_hyperlink(self, paragraph, text, url):
        """Adds a clickable hyperlink to a paragraph."""
        part = paragraph.part
        r_id = part.relate_to(url, 'http://schemas.openxmlformats.org/officeDocument/2006/relationships/hyperlink', is_external=True)

        hyperlink = OxmlElement('w:hyperlink')
        hyperlink.set(qn('r:id'), r_id)

        new_run = OxmlElement('w:r')
        rPr = OxmlElement('w:rPr')

        style = OxmlElement('w:rStyle')
        style.set(qn('w:val'), 'Hyperlink')
        rPr.append(style)

        new_run.append(rPr)
        new_run.text = text
        hyperlink.append(new_run)
        paragraph._p.append(hyperlink)


    def convert(self, markdown_text: str, output_path: str):
        """Converts Markdown text to a DOCX file."""
        self.doc = Document()
        self.create_styles()

        # Pre-process the markdown_text to remove redundant html/body tags within embedded HTML
        # This makes BeautifulSoup's job easier
        markdown_text = re.sub(r'(<div[^>]*?>)\s*<html>\s*<body>', r'\1', markdown_text, flags=re.IGNORECASE | re.DOTALL)
        markdown_text = re.sub(r'</body>\s*</html>\s*(</div>)', r'\1', markdown_text, flags=re.IGNORECASE | re.DOTALL)


        # 1. Extract Mermaid diagrams and replace them with placeholders
        mermaid_diagrams = {}
        def replace_mermaid(match):
            key = f"%%MERMAID_DIAGRAM_{len(mermaid_diagrams)}%%"
            mermaid_diagrams[key] = match.group(1).strip()
            return f"\n<p data-mermaid-placeholder='{key}'></p>\n"

        markdown_text = re.sub(r'```mermaid\n(.*?)\n```', replace_mermaid, markdown_text, flags=re.DOTALL)

        # 2. Convert the remaining Markdown to HTML
        md = markdown.Markdown(extensions=[
            'fenced_code', 'codehilite', 'tables', 'toc', 'nl2br', 'sane_lists', 'extra',
            'def_list',
            'admonition',
            'footnotes', # Added for robustness
        ])
        html = md.convert(markdown_text)

        # 3. Parse the HTML and process each element
        soup = BeautifulSoup(html, 'html.parser')

        # Iterate through direct children of the <body> or root of the parsed HTML
        for element in soup.contents:
            if isinstance(element, NavigableString):
                text_content = str(element).strip()
                if text_content:
                    p = self.doc.add_paragraph()
                    self.process_inline_code(p, text_content)
                continue

            if isinstance(element, Tag):
                # Handle mermaid placeholder specifically if it's a direct child <p>
                if element.name == 'p' and element.has_attr('data-mermaid-placeholder'):
                    key = element['data-mermaid-placeholder']
                    if key in mermaid_diagrams:
                        self.add_mermaid_diagram(mermaid_diagrams[key])
                    else:
                        print(f"Warning: Mermaid placeholder '{key}' found but no diagram code stored.")
                        self.doc.add_paragraph(f"[Mermaid placeholder missing: {key}]", style='Body Text')
                    continue

                self.process_html_element(element) # Process other HTML elements

        # 4. Save the document
        self.doc.save(output_path)
        print(f"Document successfully saved: {output_path}")

# --- Example Usage (for Colab or direct script execution) ---
if __name__ == "__main__":
    input_md_filename = '/content/output/raw_page-0002.md'
    output_docx_filename = 'output_doc.docx'

    input_md_path = Path(input_md_filename)
    output_docx_path = Path(output_docx_filename)

    output_docx_path.parent.mkdir(parents=True, exist_ok=True)

    if not input_md_path.exists():
        print(f"Error: Input file '{input_md_path}' not found.")
        print("Please upload your .md file to the Colab environment or ensure the path is correct.")
        sys.exit(1)

    try:
        with open(input_md_path, 'r', encoding='utf-8') as f:
            markdown_content = f.read()

        converter = MarkdownToDocxConverter()
        converter.convert(markdown_content, str(output_docx_path))

        print(f"\nConversion of '{input_md_path}' to '{output_docx_path}' completed successfully.")
        print(f"You can download '{output_docx_path}' from the Colab 'Files' section.")

    except UnicodeDecodeError:
        print(f"Error: Unable to decode '{input_md_path}' with UTF-8 encoding. Please check file encoding.")
        sys.exit(1)
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        sys.exit(1)
````
